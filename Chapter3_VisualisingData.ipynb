```python
from matplotlib import pyplot as plt

years = [1950,1960,1970,1980,1990,2000,2010]
gdp = [300.2,543.3,1075.9,2862.5,5979.6,10289.7,14958.3]

plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
plt.title("Nominal GDP")

plt.ylabel("Billions of $")
plt.show()
```

# Bar charts

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Ghandi", "West Side Story"]
num_oscars = [5,11,3,8,10]

# Bars are by default width 0.8. 
# Add 0.1 to the left co-ordinate to make the bar centred. 
xs = [i + 0.1 for i, _ in enumerate(movies)]

# plot bars with left x-co-ordinate [xs], heights [num_oscars]
plt.bar(xs, num_oscars)

plt.ylabel("# of Academy Awards")
plt.xlabel("Film")

# label bars with movie title at bar centres. 
plt.xticks([i+0.6 for i, _ in enumerate(movies)], movies)

plt.show()
```

```python
from collections import Counter

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]

decile = lambda grade: grade // 10 * 10

histogram = Counter(decile(grade) for grade in grades)

plt.bar([x - 4 for x in histogram.keys()],   # shift bars to the left 4
        histogram.values(),                  # bar height
        8                                    # width of 8
       )

plt.axis([-5,105,0,5])                       # x - -5 to 105; y 0 to 5

plt.xticks([10*i for i in range(11)])        # x labels at 0, 10, ..., 100
plt.xlabel("Decile")
plt.ylabel("Students")
plt.title("Distribution of Exam 1 Grades")

plt.show()
```

# Line Charts

```python
variance       = [1,2,4,8,16,32,64,128,256]
bias_squared   = [256,128,64,32,16,8,4,2,1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

# cn make multiple calls to plt.plot
# to show different things on the same chart. 
plt.plot(xs, variance, 'g-', label='variance')  # green solid line
plt.plot(xs, bias_squared, 'r-.', label='bias^2')  # red dotted line
plt.plot(xs, total_error, 'b:', label="total error")  # blude dotted line

# because we've assigned labels to the series, we get a legend for free. 
plt.legend(loc=9)
plt.xlabel("model complexity")
plt.title("The bias/ variance trade-off")
plt.show()
```

```python

friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67] 
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190] 
labels = [' a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

# label each point
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label, 
                    xy=(friend_count, minute_count),   # put label on  point
                    xytext=(5, -5),                     # offset the label slightly
                    textcoords='offset points'
                )
    
plt.title("Daily minutes vs number of friends")
plt.xlabel("Number of friends")
plt.ylabel("Daily minutes spent on site")
plt.show()
```
